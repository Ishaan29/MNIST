```python
def img_conv(path, x, y):
    
    # Convert image to greyscale
    img_conv = Image.open(path).convert('L')
    img_conv.save(r'C:\Users\Kaustubh\zzz\resources\Images\user_gray.png')
    
    # Invert
    img = Image.open(r'C:\Users\Kaustubh\zzz\resources\Images\user_gray.png')
    inverted_image = ImageOps.invert(img)
    inverted_image.save(r'C:\Users\Kaustubh\zzz\resources\Images\user_gray_inv.png')
    
    # Resize
    img = Image.open(r'C:\Users\Kaustubh\zzz\resources\Images\user_gray_inv.png')
    img_resized = img.resize((x, y), Image.ANTIALIAS)
    img_resized.save(r'C:\Users\Kaustubh\zzz\resources\Images\user_gray_inv_res.png')
    
    # Obtain pixel values
    img = Image.open(r'C:\Users\Kaustubh\zzz\resources\Images\user_gray_inv_res.png')
    arr = np.array(img)
    
    # Transform
    arr_trfm = arr / 255
    
    # Prediction
    arr_pred = arr_trfm.reshape(1, x, y, 1)
    pred = np.argmax(model.predict([arr_pred]))
    
    return pred, arr_trfm.reshape(x, y)


def trim_image(path):
    
    img = Image.open(path)
    
    bg = Image.new(img.mode, img.size, img.getpixel((0,0)))
    diff = ImageChops.difference(img, bg)
    diff = ImageChops.add(diff, diff, 2.0, -100)
    bbox = diff.getbbox()
    
    if bbox:
        return img.crop(bbox)
    
    
def crop_digits(path, n):
    
    img = trim_image(path)
    img.save(r'C:\Users\Kaustubh\zzz\resources\Images\full_crop.png')
    x, y = img.size
    
    print('You wrote: ', end='')
    
    for i in range(1, n+1):
        
        # Bbox to crop
        box = (x//n*(i-1), 0, x//n*i, y)
        
        # Crop individual digits
        img = Image.open(r'C:\Users\Kaustubh\zzz\resources\Images\full_crop.png')
        crop_img = img.crop(box)
        crop_img.save(r'C:\Users\Kaustubh\zzz\resources\Images\{}pic.png'.format(i))
        
        # Trim cropped digits
        img = trim_image(r'C:\Users\Kaustubh\zzz\resources\Images\{}pic.png'.format(i))
        img.save(r'C:\Users\Kaustubh\zzz\resources\Images\{}pic.png'.format(i*10))
        
        # Pad the trimmed digits with white space
        img = cv2.imread(r'C:\Users\Kaustubh\zzz\resources\Images\{}pic.png'.format(i*10))
        color = [255, 255, 255]
        t, b, l, r = [70, 20, 70, 70]
        new_img = cv2.copyMakeBorder(img, t, b, l, r, cv2.BORDER_CONSTANT, value=color)
        cv2.imwrite(r'C:\Users\Kaustubh\zzz\resources\Images\{}pic.png'.format(i*100), new_img)
        
        # Call maain function
        digit, pix_val = img_conv(r'C:\Users\Kaustubh\zzz\resources\Images\{}pic.png'.format(i*100), 28, 28)
        plt.imshow(pix_val, cmap='gray')
        print(str(digit), end='')
        
crop_digits(r'C:\Users\Kaustubh\zzz\resources\Images\ocr\digit_resize4.png', 1)
```
